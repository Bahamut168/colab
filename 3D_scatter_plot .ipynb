* ເອີ່ນໃຊ້ from sklearn.cluster  ແລະ import KMeans ເຂົ້າມາ

```
from sklearn.cluster import KMeans
```

* ສ້າງຕົວປ່ຽນ ax ເພື່ອເກັບການຕັ້ງຄ່າ subplot ເປັນ 1ແຖວ 1ຖັນ 1ຮູບ ແລ້ວກໍບອກວ່າ projection='3d'

* ax.scatter ຜ່ານຄ່າ x ທັ້ງ 3 ຕົວເຂົ້າໄປ x ແຕ່ລະຕົວແມ່ນແທນໃຫ້ຄ່າຂອງ ຖັນໃນຂໍ້ທີ່ 5

* ໃສ່ label X, Y, Z

```
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], alpha=.2)
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
ax.set_zlabel('X2 Label');
```

* ສ້າງໂຕປ່ຽນ model ແລະ ສະແດງຂໍ້ມູນໂຕປ່ຽນ

```
model = KMeans(n_clusters=4)
model
```

* fit model ຜົນຂອງມັນກໍຈະມາຄືກັນກັບທາງເທິ່ງ

```
model.fit(X)
```

ສະແດງ model labels ຈະເຫັນໄດ້ວ່າມັນສະແດງອອກມາ ເປັນ 4 ກຸ່ມໃນຮູບແບບ array

```
model.labels_
```

ເພີ່ມ centroid ຫຼື ຈຸດສູນກາງ ເຂົ້າໄປໃນກຣາຟ

```
model.cluster_centers_
```

( ໃຊ້ງານ feature ນີ້ບໍ່ໄດ້ ເພາະເຮັດໃນ Colab)
* ມັນເປັນການສະແດງການ Simulation ອອກມາໃນຮູບແບບ 3D ແລ້ວເຮົາສາມາດປີ້ນເບິ່ງມັນໄດ້

```
#%matplotlib qt
#%pylab qt
```

* ຈະເຫັນວ່າມັນຈະມີຈຸດເພີ່ມຂື້ນມາເຊີ່ງນັ້ນກໍຄືຈຸດ centriod

```
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], alpha=.2)
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], model.cluster_centers_[:, 2], color='orange', alpha=1, marker='D', s=70)
ax.set_xlabel('X0')
ax.set_ylabel('X1')
ax.set_zlabel('X2');
```

ເປັນການສະແດງກຣາຟອອກມາທັ້ງ 6 ຕົວຢ່າງວ່າມັນມີຈຸດ centroid 1, 2, 3, 4, 5, 6 ອັນມັນຈະເປັນແນວໃດ

```
fig = plt.figure(figsize=(18, 10))
nrows=2
ncols=3
for n in range(1, nrows * ncols + 1):
    ax=fig.add_subplot(nrows, ncols, n, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], alpha=.2)
    model = KMeans(n_clusters=n).fit(X)
    ax.scatter(model.cluster_centers_[:, 0], 
               model.cluster_centers_[:, 1], 
               model.cluster_centers_[:, 2], 
               color='orange', alpha=1, marker='D', s=70)
    ax.set_xlabel('X0 Label')
    ax.set_ylabel('X1 Label')
    ax.set_zlabel('X2 Label')
```
